```python
import pandas as pd 
import re
```

```python
# Load the CSV file using a relative path
glassdoor_df = pd.read_csv(r"../external/jobs.csv") 

glassdoor_df = glassdoor_df.drop(['Skills'], axis = 1)

glassdoor_df['Job Description'] = glassdoor_df['Job Description'].astype(str)
```

```python
from salary_processor import SalaryProcessor

glassdoor_df['Salary Estimate'] = glassdoor_df.apply(
    lambda row: SalaryProcessor.convert_salary(
        SalaryProcessor.extract_salary(row['Job Description']) 
        if pd.isna(row['Salary Estimate']) 
        else row['Salary Estimate']
    ), axis=1)

glassdoor_df['Job Title']
```

```python
# Assuming the DescriptionProcessor class has already been defined
from jobdescr_processor import DescriptionProcessor

def process_text(text):
    # Step 1: Convert text to lowercase
    text = DescriptionProcessor.text_lowercase(text)
    
    # Step 2: Remove extra whitespace
    text = DescriptionProcessor.remove_whitespace(text)
    
    # Step 3: Remove stopwords
    text = DescriptionProcessor.remove_stopwords(text)
    
    # Step 4: Sentence segmentation
    sentences = DescriptionProcessor.sentence_segmentation(text)
    
    # Step 5: Remove punctuation
    sentences_without_punctuation = [DescriptionProcessor.remove_punctuation(sentence) for sentence in sentences]
    
    return sentences_without_punctuation

# Apply the process_text function to each row of the 'text_column'
glassdoor_df['processed_text'] = glassdoor_df['Job Description'].apply(process_text)

pd.set_option('display.max_colwidth', None)

glassdoor_df['processed_text']
```
